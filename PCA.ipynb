```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
wine = pd.read_csv("wine.csv")
wine
```

```python
wine['Type'].value_counts()
```

```python
wine1 = wine.iloc[:,1:]
wine1
```

```python
wine1.info()
```

```python
#converting data
wine_ary = wine1.values
wine_ary
```

```python
#normalize data
wine_norm = scale(wine_ary)
wine_norm
```

```python
#pca impletion
pca = PCA(n_components=13)

wine_pca = pca.fit_transform(wine_norm)
wine_pca
```

```python
pca.components_
```

```python
#amount of varation in pca
var = pca.explained_variance_ratio_
var
```

```python
#cumulative variance
var1 = np.cumsum(np.round(var,4)*100)
var1
```

```python
#variance plot for pca
plt.plot(var1,color="yellow")
```

```python
#final data frame
final_vine = pd.concat([wine['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['pca1','pca2','pca3'])],axis=1)
final_vine
```

```python
#visualization of PCA
fig = plt.figure(figsize=(15,11))
sns.scatterplot(data=final_vine)
```

```python
# 1 HIERARCHICAL CLUSTERING
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,8))
dendrogram = sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
#cluster y
hclusters = AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage='ward')
```

```python
y = pd.DataFrame(hclusters.fit_predict(wine_norm),columns=['clusterid'])
y['clusterid'].value_counts()
```

```python
wine2 = wine.copy()
wine2['clusterid']=hclusters.labels_
wine2
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
  kmeans = KMeans(n_clusters=i,random_state=2)
  kmeans.fit(wine_norm)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title("elbow graph")
plt.xlabel("number of cluster")
plt.ylabel("wcss")
plt.show()
```

```python
clusters3 = KMeans(3,random_state=30).fit(wine_norm)
clusters3
```

```python
clusters3.labels_
```

```python
#assign cluster to it
wine3 = wine.copy()
wine3['clustersid']=clusters3.labels_
wine3
```

```python
wine3['clustersid'].value_counts() 
```
